# #9 The sort_values() Method
Work both on dataframe and series.
By default value comes in ascending order because it is done sort_values(ascending = True).
So for getting result in descending order we will apply sort_values(ascending = False)

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values() # By default in ascending order
```

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending =False) # In descending Order
```

```python
data.sort_values('city') # Working on dataframe , By default in ascending order
```

```python
data.sort_values('city',ascending=False)  # In decreasing order
```

### Sorting for multiple condition

```python
data.sort_values(['city','date'])    # Both in ascending Order
```

```python
 data.sort_values(['city','date'],ascending=[True,False]) # City in ascending order and date in descending order
#Abu Dhabi pahle hai lekin Abu dhabi me jo sabse last me khela gaya hai wo sabse pahle hai.
```

# #10 The drop_duplicates() Method

```python
data.drop_duplicates('city')
```

```python
data.drop_duplicates('city').shape # Number of rows has been changed
```

```python
# For multiple column:
data.drop_duplicates(['city','season'])
```

```python
data.drop_duplicates(['city','season']).shape # Functioning like a "And" operation
```

```python
# To get the winner of each season: Step-1: For all the final matches
# By default it works like keep='First' which will give the first match of the season.
# So we have used keep= 'last', so that we can get the data of last match i.e., final match.
data.drop_duplicates('season',keep='last')
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']] # Here we extract the column of winner of each season.
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```
