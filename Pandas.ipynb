# Filtering Dataframe on a condition:

```python
mask = data['city']=='Delhi'
data[mask]
```

```python
data[mask].shape
```

```python
mask1 = data['winner'] == 'Mumbai Indians'
data[mask]
```

```python
mask2 = data['season']==2010
data[mask2]
```

```python
def get_city_match_count(city):
    mask = data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count("Rajkot")
```

## For multiple condition (masking)

```python
mask1 = data['winner']== 'Mumbai Indians'
mask2 = data["season"] == 2017
mask3 = data['win_by_runs']>10
data[mask1 & mask2 & mask3]
```
